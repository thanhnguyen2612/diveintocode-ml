# Exploratory data analysis Pre-class assignment 1 Analysis of irises
## [Problem 1] Data acquisition

```python
from sklearn.datasets import load_iris
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
(X, y) = load_iris(return_X_y=True)
X = pd.DataFrame(X, columns=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'])
y = pd.DataFrame(y, columns=['Species'])
print(X.head())
print(y.head())
```

## [Problem 2] Combining data

```python
df = pd.concat([X, y], axis=1)
df.head()
```

## [Problem 3] Checking the data

```python
print("First 4 samples")
print(df.head(4))
```

```python
print("Total number of samples for each label")
print(df.Species.value_counts(ascending=True))
```

```python
print("Missing value in the feature quantity")
print(df.iloc[:,:-1].isnull().sum())
```

```python
print("Summary of 4 features")
print(df.iloc[:,:-1].describe()[1:])
```

## [Problem 4] Examining the iris dataset itself

```python
print(df.info())
```

```python
print(df[df.Species == 0].describe())
print(df[df.Species == 1].describe())
print(df[df.Species == 2].describe())
```

It seems that Species0 has smaller petal size than Species1 and Species2, while Sepal size of Species2 is bigger than the twos

## [Problem 6] Creating a diagram

```python
plt.pie(df.Species.value_counts(), autopct='%.2f', radius=2, labels=[f'Species {s}' for s in df.Species.value_counts().keys()])
plt.show()
```

```python
cols = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
data = [df[col] for col in cols]
plt.figure(figsize=(14, 6))
plt.xlabel("Feature")
plt.ylabel("Value")
plt.boxplot(data, labels=cols)
plt.show()
```

```python
plt.figure(figsize=(14, 6))
plt.violinplot(data, showmeans=True)
plt.show()
```

Violin plot shows cleared how data distributes at different quantiles

## [Problem 7] Confirming the relationship between features

```python
cols = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
fig, ax = plt.subplots(2, 3, figsize=(20, 10))
ax = ax.flatten()
idx = 0
for i in range(len(cols) - 1):
    for j in range(i + 1, len(cols)):
        ax[idx].set_xlabel(cols[i])
        ax[idx].set_ylabel(cols[j])
        ax[idx].scatter(df[df.Species == 0][cols[i]], df[df.Species == 0][cols[j]], label='Species 0')
        ax[idx].scatter(df[df.Species == 1][cols[i]], df[df.Species == 1][cols[j]], label='Species 1')
        ax[idx].scatter(df[df.Species == 2][cols[i]], df[df.Species == 2][cols[j]], label='Species 2')
        ax[idx].legend()
        idx += 1
```

```python
import seaborn as sns

sns.pairplot(df, hue='Species', palette=sns.color_palette("hls", 3))
```

```python
# Correlation coefficient matrix for 4 features
cor = df[cols].corr(method='pearson')
print(cor)
sns.heatmap(cor)
```

## [Problem 8] Explaining the results of visualized graphs and tables

From scatterplots, we can use petal_length and petal_width to differentiate Irises:

*   Species 0: petal_length around 1->2, petal_width around 0->0.7
*   Species 1: petal_length around 3->5, petal_width around 1->1.7
*   Species 2: petal_length around 5->7, petal_width around 1.4->2.5

It's evidence that petal_width and petal_length are correlated
