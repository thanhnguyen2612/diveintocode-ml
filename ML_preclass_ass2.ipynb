# Mathematics for Machine Learning Pre-Class assignment 2 Creating Dummy Data
## [Problem 1] Creating random numbers

```
import numpy as np

np.random.seed(0)

mean = [-3, 0]
cov = [[1., .8],
       [.8, 1.]]
nums = np.random.multivariate_normal(mean, cov, 500)
print("Print shape to check:", nums.shape)
```

## [Problem 2] Visualization by scatter plot

```
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(10, 6))
plt.title("View created random points using scatterplot")
plt.scatter(nums[:, 0], nums[:, 1])
plt.show()
```

## [Problem 3] Visualization by histogram

```
plt.title("Histogram of x1")
plt.xlim(-6, 3)
plt.xlabel("x1")
plt.ylabel("frequency")
plt.hist(nums[:, 0])
plt.show()

plt.title("Histogram of x2")
plt.xlim(-6, 3)
plt.xlabel("x2")
plt.ylabel("frequency")
plt.hist(nums[:, 1])
plt.show()
```

## [Problem 4] Addition of data

Create random data

```
mean = [0, -3]
cov = [[1., .8],
       [.8, 1.]]
nums_2 = np.random.multivariate_normal(mean, cov, 500)
```

Plot the random data from 2 problems on 1 scatterplot

```
combine = np.concatenate((nums, nums_2), axis=0)

plt.figure(figsize=(10, 6))
plt.title("View created random points from 2 problems")
plt.xlabel("x1")
plt.ylabel("x2")
plt.scatter(nums[:, 0], nums[:, 1], label=0)
plt.scatter(nums_2[:, 0], nums_2[:, 1], c='orange', label=1)
plt.legend()
plt.show()
```

## [Problem 5] Data combination

```
# Combine 2 ndarrays with concatenate()
combine_1 = np.concatenate((nums, nums_2), axis=0)

# Combine 2 ndarrays with vstack()
combine_2 = np.vstack((nums, nums_2))
print("They are the same:", (combine_1 == combine_2).all())
```

## [Problem 6] Labeling

```
labels = np.concatenate((np.zeros((500, 1)), np.ones((500, 1))))

labeled_data = np.concatenate((combine_1, labels), axis=1)

print("Print shape to check:", labeled_data.shape)
print(labeled_data)
```
