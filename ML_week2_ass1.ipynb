# Mathematics lesson for machine learning assignment 1 Two-dimensional array and gradient problem
## [Problem 1] Linear function

```python
import numpy as np

np.set_printoptions(suppress=True)

x = np.arange(-50, 50.1, 0.1)
y = x / 2 + 1
print(x)
print(y)
```

## [Problem 2] Array combination

```python
combine = np.concatenate((x[:, np.newaxis], y[:, np.newaxis]), axis=1)
print(combine)
print(combine.shape)
```

## [Problem 3] Find the gradient

```python
change_amount = combine[1:, :] - combine[:-1, :]
gradient = change_amount[:, 1] / change_amount[:, 0]
print(gradient)
print(gradient.shape)
```

## [Problem 4] Draw a graph

```python
import matplotlib.pyplot as plt
%matplotlib inline

# Linear function
plt.figure(figsize=(12, 6))
plt.title("y = x/2 + 1")
plt.xlabel("x")
plt.ylabel("y")
plt.plot(combine[:, 0], combine[:, 1])
plt.show()

# Gradient
plt.figure(figsize=(12, 6))
plt.title("Gradient")
plt.xlabel("x")
plt.ylabel("gradient")
plt.plot(gradient)
plt.show()
```

## [Problem 5] Python functionalization

```python
def compute_gradient(function, x_range=(-50, 50.1, 0.1)):
    """
    Calculate the gradient using the amount of change.
    Parameters
    ----------------
    function : function
      The function you want to find, the one that returns the ndarray of y when you put the ndarray of x.
    x_range : tuple
      Specify the range in the same way as the argument of np.arange ().
    Returns
    ----------------
    array_xy : ndarray, shape(n, 2)
      A combination of x and y. n depends on x_range.
    gradient : ndarray, shape(n-1,)
      Function gradient. One factor is reduced to make a difference
    """
    x = np.arange(*x_range)
    y = function(x)
    array_xy = np.concatenate((x[:,np.newaxis], y[:,np.newaxis]), axis=1)
    gradient = (array_xy[1:, 1] - array_xy[:-1, 1]) / (array_xy[1:, 0] - array_xy[:-1, 0])
    return array_xy, gradient
```

Compute gradient and draw graph: $y = x^2$ (-50 <= x <= 50)

```python
array_xy1, gradient1 = compute_gradient(lambda x: x**2)

fig, ax = plt.subplots(1, 2, figsize=(20, 6))
fig.suptitle("y = x ** 2")
ax[0].plot(array_xy1[:, 0], array_xy1[:, 1])
ax[0].set_title("Function")
ax[0].set_xlabel("x")
ax[0].set_ylabel("y")
ax[1].plot(gradient1)
ax[1].set_title("Gradient")
ax[1].set_xlabel("x")
ax[1].set_ylabel("gradient")
plt.show()
```

Compute gradient and draw graph: $y = 2x^2 + 2^x$ (-50 <= x <= 50)

```python
array_xy2, gradient2 = compute_gradient(lambda x: 2 * x**2 + 2**x)

fig, ax = plt.subplots(1, 2, figsize=(20, 6))
fig.suptitle("y = 2x**2 + 2**x")
ax[0].plot(array_xy2[:, 0], array_xy2[:, 1])
ax[0].set_title("Function")
ax[0].set_xlabel("x")
ax[0].set_ylabel("y")
ax[1].plot(gradient2)
ax[1].set_title("Gradient")
ax[1].set_xlabel("x")
ax[1].set_ylabel("gradient")
plt.show()
```

Compute gradient and draw graph: $y = sin(x^\frac{1}{2})$ (0 <= x <= 50)

```python
def func(array_x):
    return np.sin(np.sqrt(array_x))

array_xy3, gradient3 = compute_gradient(func, (0, 50.1, 0.1))

fig, ax = plt.subplots(1, 2, figsize=(20, 6))
fig.suptitle("y = sin(x**1/2)")
ax[0].plot(array_xy3[:, 0], array_xy3[:, 1])
ax[0].set_title("Function")
ax[0].set_xlabel("x")
ax[0].set_ylabel("y")
ax[1].plot(gradient3)
ax[1].set_title("Gradient")
ax[1].set_xlabel("x")
ax[1].set_ylabel("gradient")
plt.show()
```

## [Problem 6] Find the minimum value

```python
print(f"Min value of function y = x**2: {array_xy1.min(axis=0)[1]} at index {array_xy1.argmin(axis=0)[1]}")
print(f"Min value of function y = 2x**2 + 2**x: {array_xy2.min(axis=0)[1]} at index {array_xy2.argmin(axis=0)[1]}")
print(f"Min value of function y = sin(x**1/2): {array_xy3.min(axis=0)[1]} at index {array_xy3.argmin(axis=0)[1]}")
```

Slope (gradient) around the min value of the 3 functions

```python
print(gradient1[array_xy1.argmin(axis=0)[1] - 5: array_xy1.argmin(axis=0)[1] + 5])
print(gradient2[array_xy2.argmin(axis=0)[1] - 5: array_xy2.argmin(axis=0)[1] + 5])
print(gradient3[array_xy3.argmin(axis=0)[1] - 5: array_xy3.argmin(axis=0)[1] + 5])
```
